```python
import pandas as pd

"""
DataFrame：表
Series：1次元配列（縦列、横列どちらもSeriesという）
"""
```

```python
# dictionaryを使用して、pd.Seriesを作成
data = {
    "name": 'John',
    'sex': 'male',
    'age': 20,
    'height': 170,
    'weight': 70
}
John_s = pd.Series([data])
print(John_s)
```

```python
import numpy as np

# ndarrayを使用して、pd.Seriesを作成
data = np.array(['John', 'Jane', 'Mike'])
John_s = pd.Series(data)
print(John_s)
```

```python
# 第二引数にindexのラベルを設定することもできる（設定しないと、数字が割り当てられる）
data = np.array(['John', 'Jane', 'Mike'])
labels = ['a', 'b', 'c']
pd.Series(data, labels)
```

## Nan（欠損地）の処理

```python
# NaNがある行をドロップする
df.dropna()
```

```python
# NaNがあるカラムをドロップする(Age, Cabin)
df.dropna(axis=1)
```

```python
# subsetにNaNチェックの対象とする列

df.dropna(subset=['Age'])
```

```python
# NaNに特定の値を入れる
df.fillna('This is nan!')
```

```python
# 欠損値を平均値で埋める
df['Age'].fillna(df['Age'].mean())
```

```python
# 代入
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
# DataFrameの中の値のNaN判定には pd.isna() を使う
pd.isna(df)
```

```python
# NaN判定の結果を別のカラムで持つ
df['Cabin_nan'] = pd.isna(df['Cabin'])
df
```

## グループ化

```python
df[df['Pclass'] == 1]
```

```python
# groupbyを使用しない場合、以下のように記述する必要がある
df[df['Pclass']==1].describe().loc['mean']
```

## 表の結合

```python
df1 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})
 
df2 = pd.DataFrame({ 'Key': ['k0', 'k1', 'k3'],
                    'C': ['c0', 'c1', 'c3'],
                    'D': ['d0', 'd1', 'd3']})
print(df1)
print(df2)
```

```python
df1.merge(df2)
df1
```

```python
# 縦
# こっちのパターンが多い
pd.concat([df1, df2])
```

```python
# 横
pd.concat([df1, df2], axis=1)
```

```python
df1.merge(df2, how='left')
```

```python
# leftもrightも結合できるレコードは結合。できないところはNaNで埋める
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='inner')
```

```python
df1.merge(df2, on='Key')
```

```python
df1.merge(df2, on='Key', suffixes=('_left', '_right'))
```

```python
def get_age_group(age):
    return str(age)[0] + '0代'

get_age_group(28)
```

```python
df['Age_group'] = df['Age'].apply(get_age_group)
df
```

```python
# lambdaを使用
f = lambda x: str(x)[0] + '0代'
f(48)
```

```python
df['age-group'] = df['Age'].apply(lambda x: str(x)[0] + '0代')
df
```

```python
df = pd.DataFrame({ 'name': ['john', 'Mike', 'Emily'],
                    'age': ['23', '36', '42']})

df['description'] = df.apply(lambda row: '{} is {} years old'.format(row['name'], row['age']), axis=1)
df
```
