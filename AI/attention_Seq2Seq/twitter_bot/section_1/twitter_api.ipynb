# Twitter APIの使用
PythonのライブラリTweepyを使って、タイムラインへの投稿およびツイートの取得を行います。

## APIのkey
Twitter Developer PlatformからAPI Key & SecretおよびAccess Token & Secretを取得し、以下のセルの変数に設定します。
```
api_key ="API key"
api_key_secret ="API key secret"
access_token="Access token"
access_token_secret ="Access token secret"
``` 

````
api_key = ""
api_key_secret = ""
access_token= ""
access_token_secret = ""
````

## APIの設定
Tweepyを利用し、OAuth認証にアクセストークンを設定します。  
その上で、Twitterとやり取りするためのAPIを設定します。

````
!pip install tweepy==4.0.1
````

````
import tweepy

auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
````

## タイムラインへの投稿
`tweepy.Client()`によりタイムラインに投稿します。  
https://docs.tweepy.org/en/stable/client.html  
短期間に同一の内容を投稿するとブロックされるので注意しましょう。  
※2021/11/15からTwitter APIの利用が少し厳しくなりました。それにより、Twitter APIを利用したツイートをするためには、初期段階のEssentialではなく、1つ上位のElevatedに申請し承認される必要があります。

````
consumer_key = 'あなたのAPI Key'
consumer_secret = 'あなたのAPI Key Secret'
access_token = 'あなたのAccess Token'
access_token_secret = 'あなたのAccess Token Secret'
````

````
print("文章を入力してください。")
text = input()  # ユーザーの入力を取得
client = tweepy.Client(None, consumer_key, consumer_secret, access_token, access_token_secret)
client.create_tweet(text=text)
````

## Tweetの取得
`user_timeline()`によりタイムラインのツイートを取得します。  
https://kurozumi.github.io/tweepy/api.html#API.user_timeline

````
user_id = "live_ai_exp"
tweets = api.user_timeline(user_id, count=3)
for tweet in tweets:
    print("id: ", tweet.id)
    print("user: ", tweet.user.screen_name)
    print("date: ", tweet.created_at)
    print("text: ", tweet.text)
    print("fav: ", tweet.favorite_count)
    print("ret: ", tweet.retweet_count)
    print("-----------------------------------------------")
````
