## PyTorch（パイトーチ）のメリット
---
**「PyTorch」**とは、PythonのDL（ディープラーニング）用フレームワークでオープンソースで公開された。  
開発はFacebookの人工知能研究グループが行っていて、2016年に公開された比較的新しいライブラリだ。

PyTorch

#### PyTorchの対象・目的
---
公式のHPには、PyTorchの対象は次の2点であると書かれていて、
* GPUの能力を発揮するためのNumPyの代替
* 優れた柔軟性とスピードを持つディープラーニング研究のプラットフォーム    

これらを可能にする構造になっている。

#### PyTorchの特徴
---
上記の目的を可能にするPyTorchの特徴は以下の通りだ。
* GPUを利用した高速な計算が可能
* Define-by-runを採用しているためコードに柔軟性がある（後々説明する）
* Pytorch独自のデータ構造（**Tensor**：テンソル）を使うことで簡単にニューラルネットワークを作成できる
    * Tensorについては次単元から説明する
* 近年ユーザー数が最も伸びている機械学習ライブラリと言える
    
#### PyTorchでなにができるか
---
* PyTorchには、DLを構成する様々な要素が既に定義されているため、簡単にDLを組み上げることができる
* シンプルな構造ながら、カスタマイズ性が高いため複雑で難しいDLを組むこともできる
* 研究者からの人気が高く、論文で発表された最新のDL手法の実装例が手に入りやすい

## PyTorchの位置づけ
---
#### Define-and-runとDefine-by-runの違い
---
機械学習ライブラリには主に"Define-and-run"と"Define-by-run"の2種類に分かれている。その中でもpytorchはDefine-by-runの代表格だ。

- Define-and-run
    - tensorflow1系、Theano、caffe
- Define-by-run
    - pytorch、tensorflow2系、(chainer)

※ 理論的な違いの定義はやや難しいので、以下の特徴の違いだけ説明をしよう

**Define-and-run**  
計算グラフの構築した後にデータの流し込みを行う  
イメージ：モデルの骨組みを先に設計し、訓練時はデータを流し込むだけ  
- メリット
    - 演算速度が早い
    - モデルを最適化しやすい

**Define-by-run**  

計算グラフの構築と計算を同時に行う  
イメージ：訓練時にデータを流し込みながらモデルが設計される  
- メリット
    - データによって条件分岐が可能
    - 計算途中の確認が簡単で、デバッグしやすい
    - コーディングが簡潔

現在はpytorchのようなDefine-by-runが主流になりつつあり、演算速度を犠牲にして開発のしやすさが重要視されています。
