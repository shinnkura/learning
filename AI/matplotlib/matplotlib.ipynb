## subplot(画像を並べて表示)

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = x*x

# グラフを2つ並べる
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='Earning with effort')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y2, label='Earning with effort x2')
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, y1)
ax2.plot(x, y2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1)
axes[1].plot(x, y2)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
axes[1, 2].plot(x, y2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1, label='index 0 plot')
axes[1].plot(x, y2, label='index 1 plot')
axes[0].legend()
axes[1].legend()
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1, label='index 0 plot')
axes[1].plot(x, y2, label='index 1 plot')
axes[0].legend()
axes[1].legend()
```

## 保存方法

```python
# グラフのサイズを修正
fig, axes = plt.subplots(figsize=(10, 3))
axes.plot(x, y1)

# titleを設定
axes.set_title('Title')
# x軸のラベルを設定
axes.set_xlabel('X Label')
# y軸のラベルを設定
axes.set_ylabel('Y Label')

fig.savefig('test.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))
axes[0,].plot(x,y1)
axes[1,].plot(x,x*x)

axes[0,].set_title('Title')
axes[1,].set_title('Title')
axes[0,].set_xlabel('X Label')
axes[1,].set_xlabel('X Label')
axes[0,].set_ylabel('Y Label')
axes[1,].set_ylabel('Y Label')

# 重なって表示が見えずらい時に使用
plt.tight_layout()

```

```python
# pdfで保存
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages('savefig_sample.pdf')
```

```python
from matplotlib.backends.backend_pdf import PdfPages
# pdfインスタンス作成．この時点でファイルができる
pdf = PdfPages('savefig_sample.pdf')

#------グラフ作成--------
fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
axes.legend(loc=0)
#-----------------------

#pdfに保存
pdf.savefig(fig)

#close処理（一応やる）
pdf.close()
```

```python
# 大量のグラフをPDFで保存
pdf = PdfPages('savemultifig_sample.pdf')
for i in range(0, 10):
    #------グラフ作成--------
    fig, axes = plt.subplots()
    #グラフの形状が徐々にかわるように設計しました．（適当です．）
    axes.plot(x, y1 + x*i)
    #タイトルをつけます．pdfで文字検索ができることを確認してください．
    axes.set_title('ID:#{}'.format(i))
    #-----------------------

    #forループのなかで保存
    pdf.savefig(fig)

#ループ後に閉じます
pdf.close()
```
