```python
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
df = pd.read_csv('../notebooks/train.csv')
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set()
sns.distplot(df['Age'])
```

```python
# jointplot は散布図
sns.jointplot(x='Age', y='Fare', data=df)
```

```python
# kind='hex' はヒートマップ
sns.jointplot(x='Age', y='Fare', data=df, kind='hex')
```

```python
# sns.pairplot()で複数カラムの分布を一発で表示
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='Survived', kind='scatter', plot_kws={'alpha': 0.5})
# hue: 色分け
# kind: グラフの種類
# plot_kws: 散布図の透明度

# ヒストグラム
sns.distplot(df['Age'])

```

```python
sns.pairplot(df[['Age', 'Fare', 'Survived']], hue='Survived', plot_kws={'alpha': 0.5})
```

```python
# 棒グラフ
sns.barplot(x='Pclass', y='Survived', data=df)
```

```python
import numpy as np
# 平均値などの集約値を表示
sns.barplot(x='Survived', y='Age', data=df, hue='Survived', estimator=np.median)
```

```python
# xで指定したカテゴリ変数の値の出現回数（件数）を棒グラフで可視化する
sns.countplot(x='Survived', data=df, hue='Survived')
```

```python
# 箱ひげ図
sns.boxplot(x='Pclass', y='Age', data=df, hue='Pclass')
```

```python
sns.boxplot(x='Pclass', y='Age', data=df, hue='Survived')

"""
以下のグラフからわかること
- クラスが高いほど、年齢層も高い
- 同じクラスでも、若年層の生存率が高い
"""
```

```python
# データ分布の可視化
sns.violinplot(x='Pclass', y='Age', data=df, hue='Pclass')
```

```python
sns.swarmplot(x='Pclass', y='Age', data=df, hue='Pclass')
```

```python
# size：ドットの大きさ
sns.swarmplot(x='Pclass', y='Age', data=df, hue='Survived', size=5)
```

## 色付け

sns.set(context='poster')
- context: グラフのサイズ
  - paper: 小さい
  - notebook: 中くらい
  - talk: 大きい
  - poster: 特に大きい



```python
sns.set(context='paper')
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set(context='notebook')
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set(context='talk')
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set(context='poster')
df = df.dropna(subset=['Age'])
sns.distplot(df['Age'])
```

```python
sns.set(palette='bright')
sns.violinplot(x='Pclass', y='Age', data=df)
```
